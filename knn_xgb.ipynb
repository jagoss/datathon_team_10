```python
import pandas as pd
from sklearn.impute import SimpleImputer, KNNImputer

df = pd.read_csv('train.csv')

```

```python
df
```

```python
df.head()
```

```python
df.info()
```

```python
imputer = KNNImputer(n_neighbors=20)
df['edad'] = imputer.fit_transform(df[['edad']])
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns='ingreso')
y = df['ingreso']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8)
```

```python
!pip install category_encoders
```

```python

from sklearn.pipeline import Pipeline
from category_encoders.target_encoder import TargetEncoder
from xgboost import XGBRegressor

estimators = [
    ('encoder', TargetEncoder()),
    ('clf', XGBRegressor(random_state=8))
]
pipe = Pipeline(steps=estimators)
pipe
```

```python
import numpy as np
def mape_scorer(y_true, y_pred):
    """Calcula el Mean Absolute Percentage Error (MAPE)."""
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true))
```

```python
from sklearn.metrics import make_scorer

mape_scoring = make_scorer(mape_scorer, greater_is_better=False)
```

```python
!pip install scikit-optimize
```

```python

from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer

search_space = {
    'clf__max_depth': Integer(2, 14),
    'clf__learning_rate': Real(0.001, 1.0, prior='log-uniform'),
    'clf__subsample': Real(0.2, 1.0),
    'clf__colsample_bytree': Real(0.5, 1.0),
    'clf__colsample_bylevel': Real(0.5, 1.0),
    'clf__colsample_bynode': Real(0.5, 1.0),
    'clf__reg_alpha': Real(0.0, 10.0),
    'clf__reg_lambda': Real(0.0, 10.0),
    'clf__gamma': Real(0.0, 10.0),
    'clf__n_estimators': Integer(100, 500),
}

opt = BayesSearchCV(pipe, search_space, cv=3, n_iter=100, scoring=mape_scoring, random_state=8)
```

```python
opt.fit(X_train, y_train)
```

```python
print(opt.best_params_)
```

```python
opt.best_estimator_
```

```python
opt.best_score_
```

```python
opt.score(X_test, y_test)
```

```python
opt.predict(X_test)
```
